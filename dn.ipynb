```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from matplotlib.ticker import PercentFormatter
```

## Analysis of guesses (dataset includes outliers)

### Sample size = 29

```python
df = pd.read_csv('guesses.txt')
guesses = df['time']
#print(len(guesses)) (uncomment to view the sample size)
mean_gt = np.mean(guesses)
median_gt = np.median(guesses)
print("Mean guess = {:.3f} hrs, Median guess = {:.3f} hrs".format(mean_gt, median_gt))
```

```python
sns.set_style("darkgrid")  # grid
graph_dims= (7, 5)          # dimensions of the graphs
```

```python
def distribution_plotter(arr, title="title", xlabel="xlabel", mean=False, bins=10):   # function to plot distributions
    fig = plt.figure(figsize = graph_dims)
    plt.title(title)
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.hist(arr, weights = np.ones(len(arr)) / len(arr), bins=bins, color="darkcyan")
    if(mean==True):
        plt.axvline(np.mean(arr), color='red')
        plt.legend(["Mean"])
    plt.xlabel(xlabel)
    plt.ylabel("Proportion")
    plt.show()
```

```python
distribution_plotter(guesses, "Distribution of estimated reading time (with outliers)", 
"Estimated Reading time (in hrs)", bins = 20) # more bins in order to display outliers
```

## Analysis of guesses (sans outliers)

### Sample size = 23

```python
df = pd.read_csv('guesses_sans_outliers.txt')
guesses = df['time']
#print(len(guesses)) (uncomment to view the sample size)
print("Mean guess = {:.3f} hrs, Median guess = {:.3f} hrs".format(np.mean(guesses), np.median(guesses)))
```

```python
distribution_plotter(guesses, "Distribution of estimated reading time (without outliers)", "Estimated Reading time (in hrs)", bins = 8)
```
