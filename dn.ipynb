```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from matplotlib.ticker import PercentFormatter
```

```python
def distribution_plotter(arr, title="title", xlabel="xlabel", mean=False, bins=10):   # function to plot distributions
    fig = plt.figure(figsize = graph_dims)
    plt.title(title)
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.hist(arr, weights = np.ones(len(arr)) / len(arr), bins=bins, color="darkcyan")
    if(mean==True):
        plt.axvline(np.mean(arr), color='red')
        plt.legend(["Mean"])
    plt.xlabel(xlabel)
    plt.ylabel("Proportion")
    plt.show()
```

```python
sns.set_style("darkgrid") 
graph_dims= (6, 4)          # dimensions of the graphs
```
